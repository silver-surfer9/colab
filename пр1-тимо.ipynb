```python
import numpy
uniform_arr = numpy.random.uniform(-100, 51, 100)
print(uniform_arr)
```

```python
normal_arr = numpy.random.normal(10, 5, 100)
print(normal_arr)
```

```python
exp_arr = numpy.random.exponential(0.5, 100)
print(exp_arr)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.hist(uniform_arr, density=True)
plt.xlabel('Значения Х')
plt.title('Равномерное распределение')
plt.show()
plt.hist(normal_arr, density=True)
plt.xlabel('Значения Х')
plt.title('Нормальное распределение')
plt.show()
plt.hist(exp_arr, density=True)
plt.xlabel('Значения Х')
plt.title('Экспоненциальное распределение')
plt.show()
```

```python
mean_uniform = numpy.mean(uniform_arr)
std_uniform = numpy.std(uniform_arr)
z_uniform = (uniform_arr - mean_uniform) / std_uniform
print(z_uniform)
```
